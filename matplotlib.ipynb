LINE GRAPH

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10,2)
y=x**2
print(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,ls="dotted") # is means line style
plt.show()
```

```python
plt.plot(x,y,ls="dashed") # is means line style
plt.show()
```

```python
plt.plot(x,y,ls="dashdot") # ls means line style
plt.show()
```

```python
plt.plot(x,y,ls="none") # means line style
plt.show() # none represents no line
```

```python
plt.plot(x,y,marker=".",ls="none") # ls means line style
plt.show()
```

```python
plt.plot(x,y,marker=".",)
plt.show()
```

```python
plt.plot(x,y,marker=".",c="r") #c means color
plt.show()
```

```python
plt.plot(x,y,marker=".",c="k") #c means color
plt.show()
```

```python
plt.plot(x,y,marker=".",c="g") #c means color
plt.show()
```

```python
plt.plot(x,y,marker=".",c="g",lw="10") #c means color #iw means line width
plt.show()
```

```python
plt.plot(x,y,marker=".",c="b",lw="10",ls="dotted") #c means color #lw means line width
plt.show()
```

```python
x=np.arange(0,10,2)
y1=x**2
y2=x**3
y3=x**4
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show() #shows multiple lines in a single graph
```

HISTOGRAM

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="bar",rwidth=0.7)
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("no of students")
plt.show()
```

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="step",rwidth=0.7)
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("no of students")
plt.show()
```

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="stepfilled",rwidth=0.7,facecolor="r")
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("no of students")
plt.show()
```

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="bar",rwidth=0.7,facecolor="g")
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("no of students")
plt.show()
```
